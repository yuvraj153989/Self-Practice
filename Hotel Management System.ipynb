```python
class Hotel:
    def __init__(self):
        self.rooms_available = 10
        self.rooms_reserved = 0
        self.records = []

    def home_menu(self):
        while True:
            print("1. Booking")
            print("2. Enquiry")
            print("3. Exit")
            choice = int(input("How can we help you?"))
            if choice == 1:
                self.booking_menu()
            elif choice == 2:
                self.enquiry_menu()
            elif choice == 3:
                break
            else:
                print("Invalid choice. Please try again.")

    def booking_menu(self):
        while True:
            print("1. Check availability")
            print("2. Make a booking")
            print("3. Back to main menu")
            choice = int(input("How can we help you?"))
            if choice == 1:
                self.check_availability()
            elif choice == 2:
                self.make_booking()
            elif choice == 3:
                break
            else:
                print("Invalid choice. Please try again.")

    def check_availability(self):
        print("Number of rooms available:", self.rooms_available)
        print("Number of rooms reserved:", self.rooms_reserved)

    def make_booking(self):
        name = input("Enter customer name: ")
        number = input("Enter customer number: ")
        num_rooms = int(input("Enter number of rooms to book: "))
        duration = int(input("Enter stay duration: "))
        if num_rooms <= self.rooms_available:
            self.rooms_available -= num_rooms
            self.rooms_reserved += num_rooms
            record = {
                "name": name,
                "number": number,
                "num_rooms": num_rooms,
                "duration": duration
            }
            self.records.append(record)
            print("Booking successful!")
        else:
            print("Insufficient rooms available.")

    def enquiry_menu(self):
        while True:
            print("1. Search by customer name")
            print("2. Search by room number")
            print("3. Add a record")
            print("4. Edit a record")
            print("5. Delete a record")
            print("6. Back to main menu")
            choice = int(input("How can we help you?"))
            if choice == 1:
                self.search_by_name()
            elif choice == 2:
                self.search_by_room()
            elif choice == 3:
                self.add_record()
            elif choice == 4:
                self.edit_record()
            elif choice == 5:
                self.delete_record()
            elif choice == 6:
                break
            else:
                print("Invalid choice. Please try again.")

    def search_by_name(self):
        name = input("Enter customer name to search: ")
        found = False
        for record in self.records:
            if record["name"] == name:
                print("Customer Name:", record["name"])
                print("Customer Number:", record["number"])
                print("Number of Rooms:", record["num_rooms"])
                print("Duration:", record["duration"])
                found = True
                break
        if not found:
            print("No record found.")

    def search_by_room(self):
        room_num = int(input("Enter room number to search: "))
        found = False
        for record in self.records:
            if record["num_rooms"] == room_num:
                print("Customer Name:", record["name"])
                print("Customer Number:", record["number"])
                print("Number of Rooms:", record["num_rooms"])
                print("Duration:", record["duration"])
                found = True
                break
        if not found:
            print("No record found.")

    def add_record(self):
        name = input("Enter customer name: ")
        number = input("Enter customer number: ")
        num_rooms = int(input("Enter number of rooms to book: "))
        duration = int(input("Enter stay duration: "))
        record = {
            "name": name,
            "number": number,
            "num_rooms": num_rooms,
            "duration": duration
        }
        self.records.append(record)
        self.rooms_reserved += num_rooms
        print("Record added successfully!")

    def edit_record(self):
        name = input("Enter customer name to edit: ")
        found = False
        for record in self.records:
            if record["name"] == name:
                record["name"] = input("Enter new customer name: ")
                record["number"] = input("Enter new customer number: ")
                record["num_rooms"] = int(input("Enter new number of rooms to book: "))
                record["duration"] = int(input("Enter new stay duration: "))
                print("Record edited successfully!")
                found = True
                break
        if not found:
            print("No record found.")

    def delete_record(self):
        name = input("Enter customer name to delete: ")
        found = False
        for record in self.records:
            if record["name"] == name:
                self.rooms_reserved -= record["num_rooms"]
                self.records.remove(record)
                print("Record deleted successfully!")
                found = True
                break
        if not found:
            print("No record found.")



booking_system = Hotel()
booking_system.home_menu()
```
