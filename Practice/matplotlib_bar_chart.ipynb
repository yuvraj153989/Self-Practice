```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style #styling graphs,colors,background etc
```

```python
classes=['Python','ML','DL',"AI",'DS']  #available courses
class1=[30,10,20,25,10] #out of 100 students in each class
class2=[40,5,20,20,10]
class3=[35,5,30,15,15]
```

```python
plt.bar(classes,class1)  #created bar graph for classes and class1
plt.show()
```

```python
plt.barh(classes,class1)  #in horizontal manner
```

```python
plt.bar(classes,class1,width=0.3)  #can cahnge width of bars accordingly
```

```python
plt.bar(classes,class1,width=0.3,align='edge') #center se edge me ho jayega bars 
```

```python
plt.bar(classes,class1,width=0.3,align='edge',color='r') #colors
```

```python
plt.bar(classes,class1,width=0.3,align='edge',color='r',edgecolor='b',linewidth=3) #coloring edges of bar, and size change
```

```python
plt.bar(classes,class1,width=0.3,align='edge',color='r',edgecolor='b',linewidth=3,label='Class 1 Students')
```

```python
style.use("ggplot")  #bckgrnd color 
plt.title("Bar Chart of Classes",fontsize=15) #title with font size change option
plt.xlabel("Classes")
plt.ylabel("No. of Students")
plt.bar(classes,class1,width=0.3,align='edge',color='r',edgecolor='b',linewidth=3)
```

```python
style.use("ggplot")  #bckgrnd color 
plt.title("Bar Chart of Classes",fontsize=15) #title with font size change option
plt.xlabel("Classes")
plt.ylabel("No. of Students")
plt.bar(classes,class1,width=0.3,align='edge',color='r',edgecolor='b',linewidth=3)  #for class 1
plt.bar(classes,class2,width=0.3,align='edge',color='y',edgecolor='b',linewidth=3)  #for class 2
plt.bar(classes,class3,width=0.3,align='edge',color='g',edgecolor='b',linewidth=3)  #for class 3

#it'll show overlapped values
```

```python
style.use("ggplot")  #bckgrnd color 
plt.title("Bar Chart of Classes",fontsize=15) #title with font size change option
plt.xlabel("Classes")
plt.ylabel("No. of Students")
width=0.2
index=np.arange(len(classes)) #finding indices for showing values seperately
plt.bar(index,class1,width,align='edge',color='r',edgecolor='b',linewidth=3)  #for class 1
plt.bar(index+width,class2,width,align='edge',color='y',edgecolor='b',linewidth=3)  #for class 2
plt.bar(index+2*width,class3,width,align='edge',color='g',edgecolor='b',linewidth=3)  #for class 3
plt.xticks(index+width,classes)  #x axis ke upar classes ke labels

#index=0.2 for bar1
#index+width(0.2) for bar2, it'll leave 0.2 space for it
#index+2*width for bar3
```
