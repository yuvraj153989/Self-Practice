```python
classes=['ML','AI','DL','R','Python']
class1=[45,15,35,25,30]
explode=[0,0.2,0,0,0.3] #attribute will explode out from pie chart by given float %age
plt.pie(class1,labels=classes,explode=explode,colors=['c','b','r','y','g'],autopct="%0.1f%%",
       shadow=True,radius=1,textprops={"fontsize":15})  #showing in pie chart format
#colors will re define colors of different typez 
#autopct will show captured %age in pie chart,   . ke baad single value hai coz 0.1..if 0.2 hota to 2 count tak values milti
#shadow will show shadows..grey color me 
#radius will increase radius
#textprops will deal with texts
plt.show()
```
