```python
df=pd.read_csv("C:\\Users\\KIIT\\Desktop\\Play Store DataSet\\googleplaystore.csv",nrows=1000) #fetching only 1000 rows
```

```python
df.shape
```

```python
x=df['Rating'] #only taking rating and reviews as x and y variable
y=df['Reviews']
```

```python
plt.scatter(x,y) #scatter plot of x and y
```

```python
plt.scatter(x,y)
plt.title("Play Store Rating/Review Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

```python
plt.figure(figsize=[9,9]) #we can change the length and breadth of plot
plt.scatter(x,y,color='r',marker='*',s=100,alpha=0.5,edgecolor='y')
 # * marker rather than dot,s increses size,alpha transparency,edgecolor for changing edges of star
plt.title("Play Store Rating/Review Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

```python
plt.figure(figsize=[9,9]) #we can change the length and breadth of plot
plt.scatter(x,y,color='r',marker='*',s=100,alpha=0.5,edgecolor='y')
 # * marker rather than dot,s increses size,alpha transparency,edgecolor for changing edges of star

plt.scatter(x,df['Installs'],color='y',marker='o',s=100,alpha=0.5,edgecolor='c') #different scatter plot of 'Installs' 

plt.title("Play Store Rating/Review Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews and Installs")
plt.show()
```
