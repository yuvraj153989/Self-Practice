```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],'Class':[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,5],'Name':['A','B','C','D']})
df2
```

```python
pd.merge(df2,df1) #jo df pehle aayega uska column pehle print hoga
```

```python
pd.merge(df1,df2) #df1 ka column pehle, where ID is common
```

```python
pd.merge(df1,df2,on="ID") #joining on the basis of ID
```

```python
pd.merge(df1,df2,on="ID") #agar koi bhi ID same nahi hai dono DFs se...to sirf utne hi print honge jitne common hai..
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='ID',how='left') #left ka data pura fill ho jayega...
```

```python
pd.merge(df1,df2,on='ID',how='outer')  #saara data print hoga
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],'Name':['A','B','C','D']})
df2
```

```python
pd.merge(df1,df2,left_index=True,right_index=True) #sara data print hoga,no matter kuch common hai ya nai
```

```python
df1=pd.DataFrame({'ID':[5,6,7,8],'Name':['A','B','C','D']})
df1
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],'Name':['A','B','C','D']})
df2
```

```python
pd.merge(df1,df2,on="ID") #merged same data set,but differentiated as Name_x and Name_y
#this works on parameter called 'suffixes'..x and y will be replaced
```

```python
pd.merge(df1,df2,on="ID",suffixes=('_df1','_df2'))
```
