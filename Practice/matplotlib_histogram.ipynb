```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
ml_student_age=np.random.randint(18,45,(100)) #creating a random int group of student aging b/w 18-45,total 100 students
python_student_age=np.random.randint(18,60,(100)) #b/w 18-60 and total 100 students
```

```python
print(ml_student_age)
```

```python
print(python_student_age)
```

```python
plt.hist(ml_student_age,rwidth=0.8,orientation='horizontal',color='m',label='ML' #creating a histogram
plt.hist(python_student_age,rwidth=0.8,orientation='vertical',color='y',label='Python')
plt.title("ML vs Python Age group")
plt.xlabel("Student Age Category")
plt.ylabel("No. of Students")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins) #age grp of 15-20,20-25,30-35,40-45 they will be shown seperately
```

```python
plt.hist(ml_student_age,bins,rwidth=0.8) #rwidth will occupy 80% space
```
