```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\KIIT\\Downloads\\tips.csv")
```

```python
df
```

```python
sns.displot(df['tip']) #designs graph in bar plot using sns functions
plt.show()
```

```python
sns.displot(df['tip'],bins=100)  #total graph plot of 100 records
plt.show()
```

```python
from scipy.stats import norm
sns.displot(df['tip'],kde=True,rug=True,fit=norm) #kde=true will create a curve over the bars in plot
plt.show()   #rug=true will create the marks where the datas are pointed
```
