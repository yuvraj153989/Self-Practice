```python
import pandas as pd
```

```python
df=pd.DataFrame({'A':[1,2,3],'B':[10,20,30]})
display(df)
```

```python
df1=pd.DataFrame({'A':[4,5,6],'B':[40,50,60]})
display(df,df1)
```

```python
df.append(df1)
```

```python
df.append(df1,ignore_index="True",sort="False") #removing repeated indices
```
